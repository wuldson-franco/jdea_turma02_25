```python
#bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import do arquivo
kaggle_dataset_path = "/content/sample_data/State_of_data_BR_2023_Kaggle - df_survey_2023.csv"
```

```python
# Leitura do Arquivo
df = pd.read_csv(kaggle_dataset_path)
```

```python
# visualização incial dos dados
print(df.columns.tolist())
```

```python
# visualizar as 5 primeiras linhas, altere o valor de dentro dos parenteses conforme necessario
df.head()
```

```python
# Limpeza e organização das colunas
df.columns = [
    col.split(',', 1)[-1].strip()
    for col in df.columns
]
```

```python
df.head()
```

```python
df.columns = [col.replace("'", "").replace("(", "").replace(")", "").strip() for col in df.columns]
```

```python
df.head()
```

```python
# Contar linhas e colunas
print(df.shape)
```

```python
#contagem e soma dos valores nulos
print(df.isnull().sum())
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(x=df['Faixa idade'], y=df['Faixa salarial'], bins=10, kde=True)  # Substitua 'Idade' pelo nome da coluna de sua escolha
plt.title('Distribuição de Idades')
plt.show()
```

```python
categoria_mais_frequente = df['Cargo Atual'].mode()[0]  # Substitua 'Profissão' pelo nome correto da coluna
print(categoria_mais_frequente)
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(y=df['Faixa salarial'])  # Substitua 'Salário' pelo nome correto da coluna
plt.title('Boxplot do Salário')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Faixa idade'], y=df['Faixa salarial']) # Substitua pelos nomes corretos
plt.title('Dispersão entre Experiência e Salário')
plt.show()
```

```python
df = df['Quais as ferramentas/tecnologias de ETL que você utiliza no trabalho como Data Analyst?']\
        .dropna()\
        .value_counts()\
        .reset_index()

df.columns = ['ferramentas', 'contagem']

def primeiro(x):
    return x.split(',')[0]

df['primeira_ferramenta'] = df['ferramentas'].apply(primeiro)
```

```python
plt.figure(figsize=(20, 5))
sns.histplot(df['primeira_ferramenta'], bins=10, kde=True)  # Substitua 'Idade' pelo nome da coluna de sua escolha
plt.title('Distribuição de Idades')
plt.xticks(rotation=90)
plt.show()
```
